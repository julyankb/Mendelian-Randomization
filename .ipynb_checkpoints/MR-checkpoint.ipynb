```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm
```

### What does scipy.stats.norm.cdf() do?
This function returns the cumulative area to the left of a given point under a normal distribution with user-defined center and scale.
We use this function when computing p-value for MR estimate. 

```python
x = np.linspace(-4, 4, num=100)
y = norm.cdf(x, loc=0, scale=1)
plt.plot(x,y)
plt.title('Cumulative area under standard normal distribution')
plt.show()

for i in [-1.96, 1.96]:
    print('Cumulative area at %s: \t%s'%(i,norm.cdf(i)))
```

### Computing MR estimates, standard error, confidence interval and p-value for a given SNP.  
- beta_x = effect of SNP on exposure
- beta_y = effect of SNP on outcome
- se_y = standard error beta_y

```python
def mr(beta_x, beta_y, se_y):
    MR_beta = beta_y/beta_x
    MR_se = se_y/abs(beta_x) # Ask Adil why he didn't use abs() here. 
    MR_95CIL = MR_beta - 1.96 * MR_se
    MR_95CIU = MR_beta + 1.96 * MR_se
    MR_p = 2*(1-norm.cdf(abs(MR_beta)/MR_se))
    return MR_beta, MR_se, MR_95CIL, MR_95CIU, MR_p
```

### Read through aligned files, compute MR estimates, make forest plots.
- "xx.aligned.tsv" files are of format: rsid \t beta_x \t beta_y \t se_y (without spaces)

- How SNPs were aligned (manually in excel):
    - For each SNP:
        - If necessary, swap reference allele for VEGF-beta such that VEGF-beta is decreasing.
        - If necessary, swap reference allele for BMD-beta such that it matches reference allele for VEGF-beta.
            - If reference allele for BMD-beta was swapped, flip sign of BMD-beta.

```python
for filepath, region in zip(['fa.aligned.tsv','ls.aligned.tsv','fn.aligned.tsv'], ['Forearm', 'Lumbar spine', 'Femoral neck']):
    with open(filepath, 'r') as infile:
        rsid_list = []
        next(infile)
        for i, line in enumerate(infile, start=1):
            data = line.rstrip().split('\t')
            rsid, beta_x, beta_y, se_y = [data[0]] + [float(i) for i in data[1:]]
            rsid_list.append(rsid)
            MR_beta, MR_se, MR_95CIL, MR_95CIU, MR_p = mr(beta_x, beta_y, se_y)
            plt.scatter(MR_beta, i, marker='s', color='r')
            plt.plot((MR_95CIL, MR_95CIU), (i, i), 'k-', linewidth=1)
            plt.title(region)
        plt.axvline(0, color='k', linewidth=0.5)
        plt.yticks([y for y in range(1, i+1)], rsid_list)
        plt.xlabel('MR effect size for VEGF on BMD')
        plt.show()
```

### Inverse Variance Weighting
Given a sequence of observations $y_i$, with variances $\sigma^2_i$, the inverse variance weighted average (IVW) is:
$$
\hat{y} = \frac{\sum_i \frac{y_i}{\sigma^2_i}}{\sum_i \frac{1}{\sigma^2_i}}
$$

If the variances are all equal, then the IVW average becomes the simple average.

```python
for filepath, region in zip(['fa.aligned.tsv','ls.aligned.tsv','fn.aligned.tsv'], ['Forearm', 'Lumbar spine', 'Femoral neck']):
    with open(filepath, 'r') as infile:
        rsid_list = []
        MR_betas = []
        ivw_weights = []
        next(infile)
        for i, line in enumerate(infile, start=1):
            
            # Read line 
            data = line.rstrip().split('\t')
            rsid, beta_x, beta_y, se_y = [data[0]] + [float(i) for i in data[1:]]
            rsid_list.append(rsid)

            
            # Include this if u wanna exclude the following SNPs
            if rsid == 'rs10761741'\
            or rsid == 'rs2639990'\
            or rsid == 'rs4782371'\
            or rsid == 'rs6993770' :
                continue
        
            # Compute MR estimate
            MR_beta, MR_se, MR_95CIL, MR_95CIU, MR_p = mr(beta_x, beta_y, se_y)

            # For IVW
            MR_betas.append(MR_beta)            
            ivw_weights.append(MR_se**(-2))            
            
            # Plot one SNP
            plt.scatter(MR_beta, i, marker='s', color='r')
            plt.plot((MR_95CIL, MR_95CIU), (i, i), 'k-', linewidth=1)
            plt.title(region)
                    
        ivw_beta = np.dot(MR_betas, ivw_weights) / np.sum(ivw_weights)
        ivw_se = np.sum(ivw_weights)**(-1/2)

        # Compute IVW estimate and standard error
        ivw_95CIL = ivw_beta - 1.96 * ivw_se
        ivw_95CIU = ivw_beta + 1.96 * ivw_se
        
        # Plot ivw estimate
        plt.scatter(ivw_beta, i+1, marker='s', color='r')
        plt.plot((ivw_95CIL, ivw_95CIU), (i+1, i+1), 'k-', linewidth=1)
    
        plt.axvline(0, color='k', linewidth=0.5)
        plt.yticks([y for y in range(1, i+1+1)], rsid_list + ['ivw'])
        plt.xlabel('MR effect size for VEGF on BMD')
        plt.show()
```

```python
for filepath, region in zip(['fa.aligned.tsv','ls.aligned.tsv','fn.aligned.tsv'], ['Forearm', 'Lumbar spine', 'Femoral neck']):
    with open(filepath, 'r') as infile:
        rsid_list = []
        MR_betas = []
        ivw_weights = []
        next(infile)
        counter = 1
        for line in infile:
            
            # Read line 
            data = line.rstrip().split('\t')
            rsid, beta_x, beta_y, se_y = [data[0]] + [float(i) for i in data[1:]]

            # Include this if u wanna exclude the following SNPs
            if rsid == 'rs10761741'\
            or rsid == 'rs2639990'\
            or rsid == 'rs4782371'\
            or rsid == 'rs6993770' :
                continue
        
            rsid_list.append(rsid)
            counter +=1
            
            # Compute MR estimate
            MR_beta, MR_se, MR_95CIL, MR_95CIU, MR_p = mr(beta_x, beta_y, se_y)

            # For IVW
            MR_betas.append(MR_beta)            
            ivw_weights.append(MR_se**(-2))            
            
            # Plot one SNP
            #plt.scatter(MR_beta, i, marker='s', color='r')
            
            plt.plot((MR_95CIL, MR_95CIU), (counter, counter), 'k-', linewidth=1)
            
            
        plt.title(region)
        
        ivw_beta = np.dot(MR_betas, ivw_weights) / np.sum(ivw_weights)
        ivw_se = np.sum(ivw_weights)**(-1/2)

        # Compute IVW estimate and standard error
        ivw_95CIL = ivw_beta - 1.96 * ivw_se
        ivw_95CIU = ivw_beta + 1.96 * ivw_se
        
        # Plot ivw estimate
        plt.scatter(ivw_beta, 1, marker='s', color='r')
        plt.plot((ivw_95CIL, ivw_95CIU), (1, 1), 'k-', linewidth=1)
        
        plt.scatter(MR_betas, [j for j in range(counter-1)], marker='s', color='r') 
                    #s=np.dot(ivw_weights, 0.08))
        
        
        plt.axvline(0, color='k', linewidth=0.5)
        plt.yticks([y for y in range(1, counter+1+1)], ['ivw'] + rsid_list)
        plt.xlabel('MR effect size for VEGF on BMD')
        plt.show()
```

```python
class Instrumental_SNP:
  def __init__(self, rsid, beta_x, beta_y, se_y):
    self.rsid = rsid
    self.MR_beta = beta_y/beta_x
    self.MR_se = se_y/abs(beta_x)
    self.MR_95CIL = self.MR_beta - 1.96 * self.MR_se
    self.MR_95CIU = self.MR_beta + 1.96 * self.MR_se
    self.MR_p = 2*(1-norm.cdf(abs(self.MR_beta)/self.MR_se))
    #self.ivw_weight = self.MR_se**(-2)
```

```python
for filepath, region in zip(['fa.aligned.tsv','ls.aligned.tsv','fn.aligned.tsv'], ['Forearm', 'Lumbar spine', 'Femoral neck']):
    # Make list of SNP objects from file
    instrument_list = []
    invalid_snps = ['rs10761741', 'rs2639990', 'rs4782371', 'rs6993770']
    #invalid_snps = []
    with open(filepath, 'r') as infile:
        next(infile)
        for line in infile:
            data = line.rstrip().split('\t')
            rsid, beta_x, beta_y, se_y = [data[0]] + [float(i) for i in data[1:]]
            if rsid not in invalid_snps:
                instrument_list.append(Instrumental_SNP(rsid, beta_x, beta_y, se_y))
    
    # Compute IVW estimate    
    MR_betas = [snp.MR_beta for snp in instrument_list]
    ivw_weights = [snp.MR_se**(-2) for snp in instrument_list]
    ivw_beta = np.dot(MR_betas, ivw_weights) / np.sum(ivw_weights)
    ivw_se = np.sum(ivw_weights)**(-1/2)
    ivw_95CIL = ivw_beta - 1.96 * ivw_se
    ivw_95CIU = ivw_beta + 1.96 * ivw_se

    # Plot IVW estimate
    plt.scatter(ivw_beta, 0, marker='s', color='r')
    plt.plot((ivw_95CIL, ivw_95CIU), (0, 0), 'k-', linewidth=1)
    
    # Plot individual MR estimates
    for i, snp in enumerate(instrument_list, start=1):
        plt.scatter(snp.MR_beta, i, marker='s', color='r')
        plt.plot((snp.MR_95CIL, snp.MR_95CIU), (i, i), 'k-', linewidth=1)
        
   
    # Graph stuff
    plt.axvline(0, color='k', linewidth=0.5)
    plt.yticks([y for y in range(len(instrument_list)+1)], ['ivw'] + [snp.rsid for snp in instrument_list])
    plt.xlabel('MR effect size for VEGF on BMD')

    plt.show()
    
```
